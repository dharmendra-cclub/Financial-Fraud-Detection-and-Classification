```python
import numpy as np
from scipy.sparse import csr_matrix

# Transform the adjacency DataFrame to a numpy array
adj_array = df_adj.values.astype(np.float32)

# Add self-loops to the adjacency array
np.fill_diagonal(adj_array, 1)

# Compute the degree array
degree_array = np.sum(adj_array, axis=1)

# Compute the inverse square root of the degree array
degree_array_inv_sqrt = np.power(degree_array, -0.5)
degree_array_inv_sqrt[np.isinf(degree_array_inv_sqrt)] = 0

# Perform symmetric normalization
adj_array_normalized = degree_array_inv_sqrt[:, None] * adj_array * degree_array_inv_sqrt[None, :]

# Convert the normalized adjacency array to a sparse matrix
adj_sparse = csr_matrix(adj_array_normalized)

# Print the normalized adjacency matrix
print(adj_sparse.toarray())
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np

class GraphConv(nn.Module):
    def __init__(self, in_features, out_features):
        super(GraphConv, self).__init__()
        self.linear = nn.Linear(in_features, out_features)

    def forward(self, x, adj):
        x = torch.mm(adj, x)
        x = self.linear(x)
        return x

class GCN(nn.Module):
    def __init__(self, in_features, hidden_features, out_features):
        super(GCN, self).__init__()
        self.conv1 = GraphConv(in_features, hidden_features)
        self.conv2 = GraphConv(hidden_features, out_features)

    def forward(self, x, adj):
        x = F.relu(self.conv1(x, adj))
        x = F.dropout(x, training=self.training)
        x = self.conv2(x, adj)
        return x

adj_tensor = torch.from_numpy(adj_sparse.toarray()).float()
num_vertices = adj_tensor.size(0)
features = torch.ones((num_vertices, 1), dtype=torch.float)

gcn = GCN(in_features=1, hidden_features=64, out_features=num_vertices)

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

gcn = gcn.to(device)
adj_tensor = adj_tensor.to(device)
features = features.to(device)

optimizer = torch.optim.Adam(gcn.parameters(), lr=0.01)

def train_model():
    gcn.train()
    optimizer.zero_grad()
    output = gcn(features, adj_tensor)
    loss = F.mse_loss(output, adj_tensor)
    loss.backward()
    optimizer.step()
    return loss.item()

for epoch in range(2000):
    loss = train_model()
    if epoch % 200 == 0:
        print(f'Epoch: {epoch+200}, Loss: {loss}')

gcn.eval()
embeddings = gcn(features, adj_tensor)
print(embeddings)
print(embeddings.shape)
```

```python
from sklearn.manifold import TSNE
import seaborn as sns

# Convert the embeddings to a numpy array
embeddings_array = embeddings.detach().cpu().numpy()
# Perform t-SNE dimensionality reduction
tsne = TSNE(n_components=2)
embeddings_2d = tsne.fit_transform(embeddings_array)

# Plot the 2D embeddings
sns.set(rc={'figure.figsize':(8,6)})
scatter_plot = sns.scatterplot(x=embeddings_2d[:, 0], y=embeddings_2d[:, 1], color='blue', edgecolor='gray', linewidth=0.5)
scatter_plot.set(xlabel='Dimension 1', ylabel='Dimension 2', title='2D Visualization of Embeddings')
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_clusters = kmeans.fit_predict(embeddings_array)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

def plot_clusters(embeddings_2d, labels, title='2D representation of clusters', color_map=None):
    color_map = color_map or {0: 'red', 1: 'blue', 2: 'green', 3: 'yellow', 4: 'purple'}
    df = pd.DataFrame(embeddings_2d, columns=['x', 'y'])
    df['Cluster'] = labels
    plt.figure(figsize=(10, 8))
    for cluster in df['Cluster'].unique():
        cluster_data = df[df['Cluster'] == cluster]
        plt.scatter(cluster_data['x'], cluster_data['y'], c=color_map.get(cluster, 'grey'), s=80, label=f'Cluster {cluster}', edgecolors='black', lw=0.5)

    plt.title(title)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend(title='Clusters', loc='upper left')
    plt.show()
```

```python
from sklearn.decomposition import PCA

# Initialize PCA
pca_transform = PCA(n_components=2)

# Fit and transform the embeddings to 2D
embeddings_2d_transformed = pca_transform.fit_transform(embeddings_array)

# Visualize the clusters
plot_clusters(embeddings_2d_transformed, kmeans_clusters)
```
